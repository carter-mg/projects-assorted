```python
import lightkurve as lk
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

target1 = lk.search_targetpixelfile("V927 Her")[1].download()
```

```python
target1.plot(aperture_mask='pipeline')
plt.savefig("target_aperture_mask.png", dpi=300, bbox_inches='tight')
plt.close()
```

```python
lc1 = target1.to_lightcurve(method = "pld", aperture_mask = 'pipeline').normalize().remove_outliers().flatten()

period = lc1.to_periodogram("bls").period_at_max_power
```

```python
lc1.fold(period).truncate(-0.1, .1).scatter()
plt.savefig("folded_lightcurve.png", dpi=300, bbox_inches='tight')
plt.close()

lc1.plot()
plt.savefig("original_lightcurve.png", dpi=300, bbox_inches='tight')
plt.close()
```

```python
search2 = lk.search_lightcurve('V927 Her', mission = 'TESS')
search2
```

```python
sum = 0
itr = 0

for index, result in enumerate(search2):
    print(f"\n Processing: {index}")
    try:
        lc = result.download()

        if lc is not None:
            pg = lc.normalize().to_periodogram()
            sum += pg.period_at_max_power
            itr += 1
            print(f"Prominent Period: {pg.period_at_max_power}")

            ax = pg.plot()
            ax.set_xlim(2, 26)
            plt.title(f"Periodogram for {result.target}, ID: {result.dp_id}")
            #plt.savefig(f'freqplot{itr}.png')

            plt.show()
            plt.close()

        else:
            print(f"Could not download {result} ")

    except Exception as e:
        print(f"Error processing {result}: {e}")

if itr > 0:
    print(f"Finished. Avg Period = {sum / itr}")
else:
    print(f"Finished, no curves processed though.")

```
